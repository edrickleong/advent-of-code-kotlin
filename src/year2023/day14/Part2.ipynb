```kotlin
import java.io.File

val input = File("input.txt").readLines()
input
```

```kotlin
var grid = input.map { it.toList() }.toList()

val grids = linkedSetOf<List<List<Char>>>()

for (i in 0..<1_000_000_000) {
    val newGrid = grid.map { it.toMutableList() }.toMutableList()
    val rowIndices = newGrid.indices
    val colIndices = newGrid[0].indices

    for (colIndex in colIndices) {
        var lastRowIndex = 0
        for (rowIndex in rowIndices) {
            val cell = newGrid[rowIndex][colIndex]
            when (cell) {
                'O' -> {
                    newGrid[rowIndex][colIndex] = '.'
                    newGrid[lastRowIndex][colIndex] = 'O'
                    lastRowIndex++
                }

                '#' -> lastRowIndex = rowIndex + 1
            }
        }
    }

    for (rowIndex in rowIndices) {
        var lastColIndex = 0
        for (colIndex in colIndices) {
            val cell = newGrid[rowIndex][colIndex]
            when (cell) {
                'O' -> {
                    newGrid[rowIndex][colIndex] = '.'
                    newGrid[rowIndex][lastColIndex] = 'O'
                    lastColIndex++
                }

                '#' -> lastColIndex = colIndex + 1
            }
        }
    }

    for (colIndex in colIndices) {
        var lastRowIndex = rowIndices.last
        for (rowIndex in rowIndices.reversed()) {
            val cell = newGrid[rowIndex][colIndex]
            when (cell) {
                'O' -> {
                    newGrid[rowIndex][colIndex] = '.'
                    newGrid[lastRowIndex][colIndex] = 'O'
                    lastRowIndex--
                }

                '#' -> lastRowIndex = rowIndex - 1
            }
        }
    }

    for (rowIndex in rowIndices) {
        var lastColIndex = colIndices.last
        for (colIndex in colIndices.reversed()) {
            val cell = newGrid[rowIndex][colIndex]
            when (cell) {
                'O' -> {
                    newGrid[rowIndex][colIndex] = '.'
                    newGrid[rowIndex][lastColIndex] = 'O'
                    lastColIndex--
                }

                '#' -> lastColIndex = colIndex - 1
            }
        }
    }

    val lastGridIndex = grids.indexOf(newGrid)
    if (lastGridIndex != -1) {
        val loopLength = grids.size - lastGridIndex
        val loopIndex = (1_000_000_000 - 1 - i) % loopLength
        grid = grids.elementAt(lastGridIndex + loopIndex)
        break
    }

    grids.add(newGrid)
    grid = newGrid
}

grid.joinToString("\n") { it.joinToString("") }
```

```kotlin
val rowIndices = grid.indices
val colIndices = grid[0].indices

var load = 0
for (colIndex in colIndices) {
    for (rowIndex in rowIndices) {
        val cell = grid[rowIndex][colIndex]
        if (cell != 'O') continue
        val numberOfRowsToSouth = grid.size - rowIndex
        load += numberOfRowsToSouth
    }
}

load
```
